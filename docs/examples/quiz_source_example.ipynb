# Quiz Source Notebook Example

A minimal instructor source notebook demonstrating the quiz Markdown syntax.
Each cell below that contains a `#### Quiz` region is marked as a
**Manually Graded Task** cell in the nbgrader cell toolbar.

After `nbgrader generate_assignment` runs:
- Each `#### Quiz` … `#### End Quiz` block is replaced by a hidden `<span>`
  containing the base64-encoded question data.
- A new code cell calling `display_quiz("#notebook:tag")` is inserted
  immediately after.
- Students see the interactive widget but cannot read the correct answers
  from the notebook source.

This example uses the v0.3.x self-check display (correctness revealed on
answer).  For the v0.4.0+ auto-grading workflow — per-question points via
`{N}`, `graded=false` self-check mode, hidden correctness, and the
`responses.json` sidecar — see the `nb1-geography.ipynb` and
`nb2-python.ipynb` downloads, and the *Graded Quizzes* page.

**Question 1** — Single choice

#### Quiz
* (SC) "Which planet is closest to the Sun?"
  + "Mercury"          (Correct!)
  - "Venus"            (Venus is second.)
  - "Earth"            (Earth is third.)
  - "Mars"             (Mars is fourth.)
#### End Quiz

**Question 2** — Many choice

#### Quiz
* (MC) "Which of the following are prime numbers?" <2>
  + "2"
  + "3"
  - "4"   (4 = 2 × 2)
  + "5"
  - "9"   (9 = 3 × 3)
#### End Quiz

**Question 3** — Numeric (exact value)

#### Quiz
* (NM) "How many sides does a hexagon have?"
  + <6>      (Correct!)
  - (Incorrect — think about the prefix "hex-".)
#### End Quiz

**Question 4** — Numeric (range + precision)

#### Quiz
* (NM) "What is the value of π, to 2 significant figures?" [2]
  + [3.1, 3.2]   (Correct — 3.1 to 2 sig. figs.)
  - [2.9, 3.0]   (A little low.)
  - [3.2, 3.5]   (A little high.)
  - (That doesn't seem right — π ≈ 3.14159.)
#### End Quiz

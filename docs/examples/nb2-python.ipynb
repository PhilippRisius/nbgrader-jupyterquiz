# Final check — Python basics
Three modes — one graded, two ungraded.

## Self-check quiz
(`py-warmup`, `graded=false`)

Ungraded warmup — click around, get correctness feedback, no points awarded by this quiz. The task cell itself still claims 1 manual point (for the prose explanation below, in this example).

#### Quiz encoded=false graded=false
* (SC) "What is the type of `1.0`?"
  - "int"
  + "float"
  - "Decimal"
#### End Quiz

* In one sentence, explain the difference between int and float.

## Study mode
(`py-study`, `graded=false hide_correctness=true`)
No score, no correctness feedback.  Click to answer, see what you committed to, but nothing reveals whether you were right.

#### Quiz encoded=false graded=false hide_correctness=true
* (MC) "Which of these are Python built-in types? Select all that apply."
  + "list"
  + "dict"
  - "vector"
  - "hashmap"
#### End Quiz
